```python
!pip install -U transformers datasets evaluate -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from datasets import Dataset, ClassLabel
from sklearn.metrics import classification_report, confusion_matrix
from transformers import (AutoTokenizer, AutoModelForSequenceClassification,
                          TrainingArguments, Trainer)
import evaluate
```

```python
# Load data
df = pd.read_csv('Symptom2Disease.csv')
df.columns = df.columns.str.strip()

# Get all unique conditions from the dataset
df = df.dropna(subset=['review', 'condition'])
relevant_conditions = sorted(df['condition'].unique().tolist())
print(f"Found {len(relevant_conditions)} unique conditions in the dataset:")
print(relevant_conditions)

# Create label mapping
df['label'] = df['condition'].map({c: i for i, c in enumerate(relevant_conditions)})

# Split dataset
from sklearn.model_selection import train_test_split

# Check class distribution
print(f"\nClass distribution:")
print(df['condition'].value_counts())

# Ensure we have enough samples for each class for stratified split
min_samples_per_class = df['condition'].value_counts().min()
print(f"\nMinimum samples per class: {min_samples_per_class}")

if min_samples_per_class < 2:
    print("Warning: Some classes have less than 2 samples. Using random split instead of stratified.")
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
else:
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

# Convert to HuggingFace datasets
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)
class_names = relevant_conditions
train_dataset = train_dataset.cast_column('label', ClassLabel(names=class_names))
test_dataset = test_dataset.cast_column('label', ClassLabel(names=class_names))

# Evaluation metrics
metric_accuracy = evaluate.load("accuracy")
metric_precision = evaluate.load("precision")
metric_recall = evaluate.load("recall")
metric_f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": metric_accuracy.compute(predictions=preds, references=labels)['accuracy'],
        "precision": metric_precision.compute(predictions=preds, references=labels, average="weighted")["precision"],
        "recall": metric_recall.compute(predictions=preds, references=labels, average="weighted")["recall"],
        "f1": metric_f1.compute(predictions=preds, references=labels, average="weighted")["f1"]
    }

# Model evaluation pipeline
model_list = {
    # "BERT": "bert-base-uncased",
    # "BioBERT": "Rostlab/biobert-base-cased-v1",
    # "PubMedBERT": "microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract",
    # "ClinicalBERT": "emilyalsentzer/Bio_ClinicalBERT"

    "BERT": "bert-base-uncased",
    "BioBERT": "monologg/biobert_v1.1_pubmed",
    "PubMedBERT": "microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract",
    "ClinicalBERT": "emilyalsentzer/Bio_ClinicalBERT",
}

results = []

for name, model_ckpt in model_list.items():
    print(f"\n🚀 Training model: {name}")

    tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

    def tokenize_fn(examples):
        return tokenizer(examples["review"], padding="max_length", truncation=True)

    tokenized_train = train_dataset.map(tokenize_fn, batched=True)
    tokenized_test = test_dataset.map(tokenize_fn, batched=True)

    model = AutoModelForSequenceClassification.from_pretrained(model_ckpt, num_labels=len(class_names))

    training_args = TrainingArguments(
        output_dir=f"./results/{name}",
        eval_strategy="epoch",
        learning_rate=2e-5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=3,
        weight_decay=0.01,
        logging_steps=10,
        report_to="none"
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_train,
        eval_dataset=tokenized_test,
        tokenizer=tokenizer,
        compute_metrics=compute_metrics
    )

    trainer.train()
    eval_metrics = trainer.evaluate()

    # Extract the core metrics and add model name
    result_dict = {
        "model": name,
        "accuracy": eval_metrics.get("eval_accuracy", 0),
        "precision": eval_metrics.get("eval_precision", 0),
        "recall": eval_metrics.get("eval_recall", 0),
        "f1": eval_metrics.get("eval_f1", 0)
    }
    results.append(result_dict)

    # Classification report and Confusion Matrix
    preds = trainer.predict(tokenized_test)
    y_true = test_dataset['label']
    y_pred = np.argmax(preds.predictions, axis=1)

    print(f"Classification Report for {name}:\n")
    print(classification_report(y_true, y_pred, target_names=class_names))

    # For confusion matrix, use a larger figure size if there are many classes
    num_classes = len(class_names)
    fig_size = max(8, num_classes * 0.8)  # Adjust size based on number of classes

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(fig_size, fig_size))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'shrink': 0.8})
    plt.title(f"Confusion Matrix - {name}")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Compile all results
df_results = pd.DataFrame(results)
print("Results DataFrame:")
print(df_results)
print("\nDataFrame columns:", df_results.columns.tolist())

# Check if we have the expected metrics
metrics_to_plot = ['accuracy', 'precision', 'recall', 'f1']
available_metrics = [col for col in metrics_to_plot if col in df_results.columns]

if available_metrics:
    # Visualization
    df_results.set_index("model")[available_metrics].plot(kind='bar', figsize=(10,6))
    plt.title("Model Comparison on Medical Text Classification")
    plt.ylabel("Score")
    plt.ylim(0, 1)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Warning: No metrics available for plotting")
    print("Available columns:", df_results.columns.tolist())
```
